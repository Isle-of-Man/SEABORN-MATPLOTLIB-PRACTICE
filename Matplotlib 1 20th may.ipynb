```python
# multiple plots togewther on one single axes 
# gave 2 colors
# we need to provide a legend how ? plt.legend()

xval = np.arange(-3 , 3.1 , 0.1)
y1 = xval**2
y2 = xval**3
 
plt.plot(xval,y1 , label = 'square')
plt.plot(xval,y2 ,  label = 'cube')

plt.legend()
plt.show()
```

```python
# can mention legend somewhere else 
xval = np.arange(-3 , 3.1 , 0.1)
y1 = xval**2
y2 = xval**3
 
plt.plot(xval,y1)
plt.plot(xval,y2)

plt.legend(['square' , 'cube'])
plt.show()
```

```python
# change the position of legend
xval = np.arange(-3 , 3.1 , 0.1)
y1 = xval**2
y2 = xval**3
 
plt.plot(xval,y1)
plt.plot(xval,y2)

plt.legend(['square' , 'cube'] , loc = 'lower right')
plt.show()
```

```python
# remove legend border 

xval = np.arange(-3 , 3.1 , 0.1)
y1 = xval**2
y2 = xval**3
 
plt.plot(xval,y1)
plt.plot(xval,y2)

plt.legend(['square' , 'cube'] , loc = 'lower right' , frameon = False)
plt.show()
```

```python
# index gets clustered so decluster the legend 
xval = np.arange(-3 , 3.1 , 0.1)
y1 = xval**2
y2 = xval**3
 
plt.plot(xval,y1)
plt.plot(xval,y2)

plt.legend(['square' , 'cube'] , ncol = 2 , loc = 'lower right')
plt.show()
```

```python
# plotting the intersection 
in our case (0,0) and (1,1)


xval = np.arange(-3 , 3.1 , 0.1)
y1 = xval**2
y2 = xval**3
 
plt.plot(xval,y1)
plt.plot(xval,y2)

plt.text(0,0 , "(0,0)")
plt.text(1,1 , "(1,1)")
plt.legend(['square' , 'cube'] , loc = 'lower right')
plt.show()
```

```python
# adding grid 

xval = np.arange(-3 , 3.1 , 0.1)
y1 = xval**2
y2 = xval**3
 
plt.plot(xval,y1)
plt.plot(xval,y2)

plt.grid()

plt.legend(['square' , 'cube'] , ncol = 2 , loc = 'lower right')
plt.show()
```

```python
# subplot 

x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)

# plt.plot(x,y2)
# plt.plot(x,y2)  not wanna plot it together 

fig , ax = plt.subplots(3,1) # 3 row and 1 col 
# ax = list and has reference to diff axises = (3,1) 
ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
```

```python
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x) 

plt.figure()

plt.subplot(2 , 3, 1) # 2 row 3 col and 1st pos 
plt.plot(x,y1)

plt.subplot(2 , 3, 3) # 2 row 3 col and 3rd pos 
plt.plot(x,y2)

plt.subplot(2 , 3, 5) # 2 row 3 col and 1st pos 
plt.plot(x,y3)



# created 2 rows and 3 cols 
# 1 3 5 is the positiom of graph and it starts from 1 and not 0
```
